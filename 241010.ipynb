```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

x=np.array([168,170,180,178,165,185,176,175,169,160]) # 아버지 키 정보
y=np.array([110,125,130,120,125,135,130,115,119,115]) # 아들 키 정보

plt.scatter(x, y)
plt.show()

corr, pvalue = stats.pearsonr(x, y)
print(corr, pvalue) # 상관계수, 유의확률
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

buy_data = [['apple','egg','milk','fish'],
            ['banana','nut','meat','milk','beer'],
             ['apple','milk','egg','fish'],
              ['apple','meat','fish','egg'],
               ['orange','milk','meat','egg']]

te = TransactionEncoder() # 머신러닝 하기 좋은 포맷으로 전환
te_arr = te.fit(buy_data).transform(buy_data)
df = pd.DataFrame(te_arr, columns=te.columns_) # one-hot 인코딩 배열로 변환
df
```

```python
freq_items = apriori(df, min_support=0.4, use_colnames=True) # 0.4이상의 support만 포함
freq_items.sort_values('support', ascending=False)
```

```python
freq_items['length'] = freq_items['itemsets'].apply(lambda x: len(x)) # 길이 열 추가
freq_items1 = freq_items[(freq_items['length']==2) & (freq_items['support']>=0.6)]

freq_items
freq_items1
```

```python
rules1=association_rules(freq_items, metric='lift', min_threshold=1.2)
rules1
```
