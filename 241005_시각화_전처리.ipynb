```python
# 연습문제
import pandas as pd
pd.options.display.max_columns=10 # 한 화면에 최대 몇개의 column까지 나타낼지
pd.set_option('display.width',1000)
stock=pd.read_csv('/content/drive/MyDrive/pythondataanalysis/stock.csv')

stock.head()
stock.tail()
```

```python
# 연습문제 1
stock.drop(index=22)
```

```python
# 연습문제 2
item = ['XOM','EXXON MOBIL CORPORATION',97.59,25588794,411102100000,16.0,97.9300]
stock.loc[22]=item

stock.tail()
```

```python
# 연습문제 3
stock=stock.sort_values(by='Volume', ascending=False) # volume 열을 기준으로 내림차순 정렬
stockpop = stock[['Symbol','Volume']] # symbol과 volume 열만 pop
stockpop.head(10)
```

```python
# 연습문제 4

stock['52_high(%)']=(stock['Price']-stock['52_high'])/stock['52_high']
# 새로운 열 생성

stock=stock.sort_values(by='52_high(%)', ascending=False)
# 52주 고점 대비 현재 가격의 하락률 열을 기준으로 내림차순으로 정렬 (큰 것부터 작은 순으로)

stock.head(10)

stockpop = stock[['Symbol','Volume','52_high(%)']]
stockpop.head(10)
```

```python
# 연습문제 5
condition = stock[stock['P/E']<20]
condition = condition[condition['52_high(%)']>-0.15]

print(condition)
```

```python
# 연습문제 6
condition2 = stock[stock['Volume']>25000000]
condition2 = condition2[condition2['Price']<150]

print(condition2)
```
