```python
# 구구단 출력
first = np.arange(2,10) # 단 2,3,4,5,6,7,8,9
second = np.arange(1,10) # 1,2,3,4,5,6,7,8,9
second = second.reshape(9,1)

result = np.zeros((len(first),len(second)),dtype=int)

# 구구단 결과를 result 배열에 저장
for i, x in enumerate(first):
    for j, y in enumerate(second):
        result[i, j] = x * y.item()

for i, x in enumerate(first):
    for j, y in enumerate(second):
        print(f"{x} * {y.item()} = {result[i, j]}")
```

```python
grade = np.zeros((4,3))
for i in range(4):
  for j in range(3):
    grade[i,j] = np.random.randint(1,100) # 1~99

print(grade)

stmean = np.zeros(4)
for i in range(4):
  stmean[i] = np.mean(grade[i,:])

print("student mean",stmean)

submean = np.zeros(3)
for j in range(3):
  submean[j] = np.mean(grade[:,j])

print("subject mean",submean)

submax = np.zeros(3)
for j in range(3):
  submax[j] = np.max(grade[:,j]) # 각 열에서 최고 점수

print("subject max grade",submax)
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

buy_data = [['apple','egg','milk','fish'],
['banana','nut','meat','milk','beer'],
['apple','milk','egg','fish'],
['apple','meat','fish','egg'],
['orange','milk','meat','egg']]

te = TransactionEncoder()
te_arr = te.fit(buy_data).transform(buy_data)
df = pd.DataFrame(te_arr, columns = te.columns_)
df
```

```python
freq_items = apriori(df,min_support=0.4,use_colnames=True)
#freq_items.sort_values('support',ascending=False)
print(freq_items)
```

```python
freq_items['length']=freq_items['itemsets'].apply(lambda x:len(x))
freq_items1 = freq_items[(freq_items['length']==2)&(freq_items['support']>=0.6)]
freq_items
freq_items1
```

```python
rules1 = association_rules(freq_items, metric='lift', min_threshold=1.2)
print(rules1)
```
