```python
import numpy as np
import pandas as pd

pop = pd.read_csv('/content/drive/MyDrive/population.csv')
print(pop)
```

```python
print(pop.describe())
```

```python
print(pop["지역"])
print(pop["인구"])
```

```python
print("population max",pop["인구"].max())
print("population min",pop["인구"].max())
print("area max",pop["면적"].max())
print("area min",pop["면적"].min())
print("density max",pop["인구밀도"].max())
print("density min",pop["인구밀도"].min())
```

```python
gu_pop=pop[["지역","인구"]]
gu_area_density=pop[["지역","면적","인구밀도"]]
print(gu_pop.head(10))
print(gu_area_density.head(10))
```

```python
pop_high = pop[pop["인구"]>500000]
pop_low = pop[pop["인구"]<300000]
print(pop_high)
print(pop_low)
```

```python
pop_high = pop[pop["인구"]>400000]
pop_high = pop_high[pop_high["면적"]<30]
print(pop_high)
```

```python
pop_gu1 = pop.loc[pop["인구"]>500000]
pop_gu2 = pop.loc[pop["인구"]>500000,"지역"]
print(pop_gu1)
print(pop_gu2)
```
